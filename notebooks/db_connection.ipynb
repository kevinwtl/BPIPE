```python
from subscription import BPIPE
import pandas as pd
x = BPIPE(['3738 HK Equity','970 HK Equity'])
x.subscribe()
```

```python
for msg in x.msgList:
    if 'MBO_TABLE_ASK' in msg and 'MBO_TABLE_BID' in msg and msg['EID'] == 14112:
        MBO_TABLE_ASK = msg.pop('MBO_TABLE_ASK')
        MBO_TABLE_BID = msg.pop('MBO_TABLE_BID')
        df_to_write = pd.concat([pd.DataFrame(msg,[0]), pd.DataFrame(MBO_TABLE_ASK).assign(MKTDEPTH_EVENT_SUBTYPE = 'INIT_PAINT_TOP'), pd.DataFrame(MBO_TABLE_BID).assign(MKTDEPTH_EVENT_SUBTYPE = 'INIT_PAINT_TOP')], ignore_index=True)
    elif 'MBO_TABLE_ASK' in msg and 'MBO_TABLE_BID' in msg and msg['EID'] == 53714:
        MBO_TABLE_ASK = msg.pop('MBO_TABLE_ASK')
        MBO_TABLE_BID = msg.pop('MBO_TABLE_BID')
        df_to_write = pd.concat([pd.DataFrame(msg,[0]), pd.DataFrame(MBO_TABLE_ASK).assign(MKTDEPTH_EVENT_SUBTYPE = 'INIT_PAINT_MBO'), pd.DataFrame(MBO_TABLE_BID).assign(MKTDEPTH_EVENT_SUBTYPE = 'INIT_PAINT_MBO')], ignore_index=True)
    elif 'MBL_TABLE_ASK' in msg and 'MBL_TABLE_BID' in msg and msg['EID'] == 14112:
        MBL_TABLE_ASK = msg.pop('MBL_TABLE_ASK')
        MBL_TABLE_BID = msg.pop('MBL_TABLE_BID')
        df_to_write = pd.concat([pd.DataFrame(msg,[0]), pd.DataFrame(MBL_TABLE_ASK).assign(MKTDEPTH_EVENT_SUBTYPE = 'INIT_PAINT_MBL'), pd.DataFrame(MBL_TABLE_BID).assign(MKTDEPTH_EVENT_SUBTYPE = 'INIT_PAINT_MBL')], ignore_index=True)
    else:
        df_to_write = pd.DataFrame(msg,[0])

    print('success')

    df_to_write.to_sql(name="test", con=x.db_connection, if_exists="append", index=False)
```
